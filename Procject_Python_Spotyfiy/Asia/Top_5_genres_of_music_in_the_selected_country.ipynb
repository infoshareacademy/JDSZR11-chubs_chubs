Wybr kraju dla ktrego stworzymy wykres top 5 gatunkw muzycznych

```python
# Wybr kraju wprowadzonego przez uytkownika
Wprowad_kraj = input("Wprowad nazw kraju, dla ktrego chcesz zobaczy top 5 gatunkw muzycznych: ")
```

```python
# Wybr tylko danych dotyczcych wybranego kraju i gatunku muzycznego
nowy_fd = fd.loc[fd['Country'] == Wprowad_kraj, ['Genre', 'Country']]
```

```python
nowy_fd.head(5)
```

```python
# Grupowanie po kraju i gatunku muzycznym, a nastpnie obliczanie liczby wystpie

zliczanie = nowy_fd['Genre'].value_counts() #zlicza nam ile jest wystpie danej kategorii
zliczanie.columns = ['Genre', 'Counts'] #to zmienia nazwe kolumn na genre i counts 
```

```python
zliczanie
```

```python
sprawdzenie = nowy_fd['Genre'].value_counts().nlargest(5)
```

```python
sprawdzenie
```

```python
top5_counts = nowy_fd['Genre'].value_counts().nlargest(5).reset_index().rename(columns={'Genre': 'Counts', 'index': 'Genre'})
```

```python
top5_counts
```

```python
# Tworzenie wykresu supkowego

plt.figure(figsize=(10,6))
plt.bar(top5_counts['Genre'], top5_counts['Counts'])

plt.title(f"Top 15 gatunkw muzycznych wedug popularnoci - {Wprowad_kraj}")

plt.xlabel("Gatunki muzyczne")
plt.ylabel("Liczba wystpie")

plt.xticks(rotation=90)
plt.show()
```
