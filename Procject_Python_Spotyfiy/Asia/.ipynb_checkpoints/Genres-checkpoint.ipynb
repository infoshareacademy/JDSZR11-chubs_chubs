```python
#Wczytanie bazy danych z wybranymi kolumnami
fd = pd.read_csv('Final database.csv', usecols=['Country', 'Popularity', 'Genre'])
```

```python
#Wyświetlenie 20 przykładowych wierszy
fd.sample(20)
```

Sprawdzenie danych

```python
#Sprawdzanie ile jest wartości:

fd.count()
```

```python
#Ile jest unikalnych gatunków muzycznych w naszej bazie?

unique_genre_count = fd['Genre'].nunique()
unique_genre_count
```

```python
#Ile jest wystąpień równych gatunków muzycznych łącznie w danym kraju?

count_genre2 = fd.groupby('Country')['Genre'].nunique()
count_genre2
```

```python
# W jakim kraju jest największe zróżnicowanie muzyczne? Top 5

unique_genre_counts = fd.groupby('Country')['Genre'].nunique().nlargest(5)
unique_genre_counts
```

```python
plt.title('Top 5 krajów z największym zróżnicowaniem muzycznym ')
plt.xlabel('Kraj')
plt.ylabel('Liczba różnych gatunków')

unique_genre_counts = fd.groupby('Country')['Genre'].nunique().nlargest(5)
plt.bar(unique_genre_counts.index, unique_genre_counts.values)
```

```python
#Jaki gatunek jest najczęściej słuchany w danym kraju? Licza wystąpień
result = fd.groupby('Country')['Genre'].apply(lambda x: x.value_counts().nlargest(1))
result
```
